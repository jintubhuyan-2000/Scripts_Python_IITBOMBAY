```python
import numpy as np
import rasterio
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# File paths
burn_area_path = r"D:\Book Chapter _ Elsvier\Burn Area\Kandhamal\Clipped\dNBR_Clip1.tif"
carbon_loss_path = r"D:\Book Chapter _ Elsvier\Carbon\Clipped\Kandhamal2023Carbon_AGB_Clip1.tif"

# Read the burn area raster
with rasterio.open(burn_area_path) as src1:
    burn_area = src1.read(1)

# Read the carbon loss raster
with rasterio.open(carbon_loss_path) as src2:
    carbon_loss = src2.read(1)

# Ensure the rasters have the same shape
if burn_area.shape != carbon_loss.shape:
    raise ValueError("The input rasters must have the same shape")

# Flatten the arrays and remove NaN values
burn_area_flat = burn_area.flatten()
carbon_loss_flat = carbon_loss.flatten()

# Mask NaN values
mask = ~np.isnan(burn_area_flat) & ~np.isnan(carbon_loss_flat)
burn_area_clean = burn_area_flat[mask]
carbon_loss_clean = carbon_loss_flat[mask]

# Calculate the correlation
correlation = np.corrcoef(burn_area_clean, carbon_loss_clean)[0, 1]

# Fit a linear regression model
model = LinearRegression()
model.fit(burn_area_clean.reshape(-1, 1), carbon_loss_clean)
r2_value = model.score(burn_area_clean.reshape(-1, 1), carbon_loss_clean)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(burn_area_clean, carbon_loss_clean, alpha=0.5, edgecolors='k', label=f'Correlation: {correlation:.2f}')
plt.plot(burn_area_clean, model.predict(burn_area_clean.reshape(-1, 1)), color='red', label=f'R²: {r2_value:.2f}')
plt.xlabel('Burn Area (dNBR)')
plt.ylabel('Carbon Loss (AGB)')
plt.title('Pixel-wise Correlation between Burn Area and Carbon Loss')
plt.legend()
plt.grid(True)
plt.show()

print(f"Pixel-wise correlation: {correlation}")
print(f"R² value: {r2_value}")
```

```python
import numpy as np
import rasterio
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# File paths
burn_area_path = r"D:\Book Chapter _ Elsvier\Burn Area\Kandhamal\Clipped\dNBR_Clip1.tif"
carbon_loss_path = r"D:\Book Chapter _ Elsvier\Carbon\Clipped\Kandhamal2023Carbon_AGB_Clip1.tif"

# Read the burn area raster
with rasterio.open(burn_area_path) as src1:
    burn_area = src1.read(1)

# Read the carbon loss raster
with rasterio.open(carbon_loss_path) as src2:
    carbon_loss = src2.read(1)

# Ensure the rasters have the same shape
if burn_area.shape != carbon_loss.shape:
    raise ValueError("The input rasters must have the same shape")

# Flatten the arrays and remove NaN values and values from burn area > 500
burn_area_flat = burn_area.flatten()
carbon_loss_flat = carbon_loss.flatten()

# Mask NaN values and burn area values > 500
mask = ~np.isnan(burn_area_flat) & ~np.isnan(carbon_loss_flat) & (burn_area_flat <= 500)
burn_area_clean = burn_area_flat[mask]
carbon_loss_clean = carbon_loss_flat[mask]

# Take a random sample of 500 points
if len(burn_area_clean) > 500:
    sample_indices = np.random.choice(len(burn_area_clean), 500, replace=False)
    burn_area_sample = burn_area_clean[sample_indices]
    carbon_loss_sample = carbon_loss_clean[sample_indices]
else:
    burn_area_sample = burn_area_clean
    carbon_loss_sample = carbon_loss_clean

# Calculate the correlation
correlation = np.corrcoef(burn_area_sample, carbon_loss_sample)[0, 1]

# Fit a linear regression model
model = LinearRegression()
model.fit(burn_area_sample.reshape(-1, 1), carbon_loss_sample)
r2_value = model.score(burn_area_sample.reshape(-1, 1), carbon_loss_sample)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(burn_area_sample, carbon_loss_sample, alpha=0.5, edgecolors='k', label=f'Correlation: {correlation:.2f}')
plt.plot(burn_area_sample, model.predict(burn_area_sample.reshape(-1, 1)), color='red', label=f'R²: {r2_value:.2f}')
plt.xlabel('Burn Area (dNBR)')
plt.ylabel('Carbon Loss (AGB)')
plt.title('Pixel-wise Correlation between Burn Area and Carbon Loss (Sampled)')
plt.legend()
plt.grid(True)
plt.show()

print(f"Pixel-wise correlation: {correlation}")
print(f"R² value: {r2_value}")
```
