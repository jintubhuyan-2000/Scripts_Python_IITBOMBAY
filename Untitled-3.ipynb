```python
import pandas as pd
import plotly.express as px

# Load your CSV data
file_path = "D:/IIT_BOMBAY/October/MHA_DISTRICT_WISE/CSV/fire event monthly.csv"
df = pd.read_csv(file_path)

# Reshape the dataframe to have columns 'District', 'Year', and 'FireEvents'
df = df.melt(id_vars=['District'], var_name='Year', value_name='FireEvents')

# Convert 'Year' to an integer, ignoring non-numeric year entries
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df.dropna(subset=['Year'], inplace=True)  # Remove rows where Year is NaN
df['Year'] = df['Year'].astype(int)  # Ensure 'Year' is integer

# Create a complete range of years (e.g., 2005 to 2023)
min_year = 2005
max_year = 2023
all_years = pd.DataFrame({'Year': range(min_year, max_year + 1)})

# Merge the complete range of years with each district to ensure all years are included
df_complete = df.merge(all_years, on='Year', how='right', sort=True)

# Fill missing 'District' and 'FireEvents' data appropriately
df_complete['District'].fillna(method='ffill', inplace=True)  # Forward fill District names
df_complete['FireEvents'].fillna(0, inplace=True)  # Fill missing fire events with 0

# Sort the dataframe by 'District' for better clarity
df_complete = df_complete.sort_values("District", ascending=True)

# Create the scatter plot with text labels showing the number of fire events
fig = px.scatter(
    df_complete, y='Year', x='District', color='FireEvents', size='FireEvents', size_max=200,
    text='FireEvents',  # Add text labels to display the number of fire events
    color_continuous_scale=px.colors.sequential.Jet, 
    labels={"FireEvents": "Fire Events"}
)

# Customize layout
fig.update_layout(
    paper_bgcolor="white",
    plot_bgcolor="white",
    height=1300, width=1200,
    title="Forest Fire Events Counts Yearly by District and Year (2005 - 2023)",
    title_x=0.5,
    xaxis=dict(tickmode='linear', dtick=1),  # Continuous ticks for Districts
    yaxis=dict(tickmode='linear', dtick=1),  # Continuous ticks for Years
)

# Add gridlines for better readability
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGray')
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGray')

# Customize colorbar
fig.update_coloraxes(colorbar=dict(title='Fire Events'))

# Adjust marker sizing for better visualization
fig.update_traces(marker=dict(sizeref=1.8), textposition='middle center')  # Set text position to center

# Show the legend
fig.update_layout(showlegend=True)

# Show the plot
fig.show()
```
